## Machine Learning

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
```

#### Logistic Regression

```python
max_accuracy = 0
r_stat = 0

for i in range (200):

    x_train, x_test, y_train, y_test = train_test_split(x, y , test_size = 0.30, random_state = i)
    model = LogisticRegression()
    model.fit(x_train, y_train)
    m_pred = model.predict(x_test)
    accurancy = accuracy_score(y_test, m_pred)
#     print(r*100, "at random-state", i)  
    if accurancy > max_accuracy:
        max_accuracy = accurancy
        r_stat = i
print("Best Accurancy is ",max_accuracy , "at  random state",  r_stat)
# Best Accurancy is  0.9679207920792079 at  random state 180
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y , test_size = 0.30, random_state = r_stat)

logistic_r = LogisticRegression()
logistic_r.fit(x_train, y_train)
l_predict  = logistic_r.predict(x_test)

print(accuracy_score(y_test, l_predict))
print(classification_report(y_test, l_predict))
print(confusion_matrix(y_test, l_predict))

# accuracy: 0.9679207920792079
```

```python
cross_val_score(logistic_r, x, y, cv = 5).mean()
# cvs = 0.8364066824455255
```

#### Decision Tree Classifier

```python
dtc = DecisionTreeClassifier()
dtc.fit(x_train, y_train)
dtc_predict = dtc.predict(x_test)

print(accuracy_score(y_test, dtc_predict))
print(classification_report(y_test, dtc_predict))
print(confusion_matrix(y_test, dtc_predict))

# Accuracy : 1.0
```

```python
cross_val_score(dtc, x, y, cv = 5).mean()
# cvs : 0.9358288770053477
```

#### Support Vection Classifier

```python
svc = SVC()
svc.fit(x_train, y_train)
svc_predict = svc.predict(x_test)

print(accuracy_score(y_test, svc_predict))
print(classification_report(y_test, svc_predict))
print(confusion_matrix(y_test, svc_predict))

# Accuracy: 0.9912871287128713
```

```python
cross_val_score(svc, x, y, cv = 5).mean()
# 0.8553964967546077
```

#### RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(x_train, y_train)
rfc_predict = rfc.predict (x_test)


print(accuracy_score(y_test, rfc_predict))
print(classification_report(y_test, rfc_predict))
print(confusion_matrix(y_test, rfc_predict))
# accuracy: 1.0
```

```python
cross_val_score(rfc, x, y, cv = 5).mean()

# accuracy :0.9334521687462864
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameter = { 'max_depth' : np.arange(2, 10), 
            'criterion' : ['gini', 'entropy'],
             'max_features' : ['auto', 'sqrt', 'log2'],
            'splitter': ['best', 'random']}
```

```python
gcv = GridSearchCV(DecisionTreeClassifier(), parameter, cv =5)
gcv.fit(x_train, y_train)
```

```python
gcv.best_params_
```

```python
final_model = DecisionTreeClassifier(criterion = 'entropy', max_depth = 9, max_features = 'auto', splitter = 'best' )
final_model.fit(x_train, y_train)
final_predict = final_model.predict(x_test)
```

```python
print(accuracy_score(y_test, final_predict))
print(classification_report(y_test, final_predict))
print(confusion_matrix(y_test, final_predict))
```
