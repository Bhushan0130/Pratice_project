## WebScraping List of airports by IATA airport code

```python
# Imported required libraries 

import selenium                             # for Webdriver
import pandas as pd                         # for DataFrame
from selenium import webdriver              # Driver
import warnings                             # To avoid any warning message
warnings.filterwarnings('ignore')           
import string                               # To get alphabeticals
```

```python
# Working to get all urls from which we have to fetch data

urls = []
path = 'https://en.wikipedia.org/wiki/List_of_airports_by_IATA_airport_code:_'
for letter in string.ascii_uppercase:
    urls.append(path+letter)
    
```

```python
# Function will work to location the element and will scrap

def data_scraping(url):
    driver = webdriver.Chrome('/Users/bhushansharma/chromedriver_mac64/chromedriver')
    driver.get(url)
    
    # Location of elements
    search_table = driver.find_elements("xpath",'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td')
    search_header = driver.find_elements("xpath", '//table[@class="wikitable sortable jquery-tablesorter"]/thead/tr/th')
    
    
    # Exception for Page Q where, two last row cell was merge
    if url[-1] == 'Q':
        for i in range(0,(len(search_table)-8),6):
            IATA.append(search_table[i].text)
            ICAO.append(search_table[i+1].text)
            Airport_name.append(search_table[i+2].text)
            Location_served.append(search_table[i+3].text)
            Time.append(search_table[i+4].text)
            DST.append(search_table[i+5].text)
            Header.append(url[-1])
            
    # Exception for few pages where data pattern is changing 
    elif len(search_header) == 5:
        for i in range(0,len(search_table),4):
            IATA.append(search_table[i].text)
            ICAO.append(search_table[i+1].text)
            Airport_name.append(search_table[i+2].text)
            Location_served.append(search_table[i+3].text)
            Time.append("")
            DST.append("")
            Header.append(url[-1])
        
    # Scraping data from pages
    else: 
        
        for i in range(0,len(search_table),6):
            IATA.append(search_table[i].text)
            ICAO.append(search_table[i+1].text)
            Airport_name.append(search_table[i+2].text)
            Location_served.append(search_table[i+3].text)
            Time.append(search_table[i+4].text)
            DST.append(search_table[i+5].text)
            Header.append(url[-1])
            
```

```python
IATA = []
ICAO = []
Airport_name = []
Location_served = []
Time = []
DST = []
Header = []

for url in urls:
    data_scraping(url)   # Function calling
```

```python
# saving DataFrame in CSV file format
data_df.to_csv('raw_data.csv')
```
