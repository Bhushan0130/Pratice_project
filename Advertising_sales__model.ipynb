## Removing Multicollinearity

#### Using VIF (Variance Inflation factor)

```python
plt.figure(figsize = (5, 3))
sns.heatmap(df.corr(), annot = True, cmap = 'Blues')   # Again draw it just for reference, Already explain in the above heatmap
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def vif_cal(data):
    vif = pd.DataFrame()
    vif['Features'] = x.columns
    vif['VIF'] = [variance_inflation_factor(x.values, i) for i in range(data.shape[1])]
    return vif
```

```python
vif_cal(x)

# Features	VIF
# 0	TV	3.049871
# 1	radio	3.587039
# 2	newspaper	4.758310
```

```python
# VIF value of each column are under VIF acceptable range
```

## Apply Machine Learning 

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from sklearn.model_selection import cross_val_score
```

## 1. LinearRegression

```python
for i in range(0, 200 ) :
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.30, random_state = i)
    model1 = LinearRegression()
    model1.fit(x_train, y_train)
    model1_predict_train = model1.predict(x_train)
    model1_predict_test = model1.predict(x_test)
    
    print('At random state', i, 'the training accuracy is :-', r2_score(model1_predict_train, y_train))
    print('At random state', i, 'the training accuracy is :-', r2_score(model1_predict_test, y_test)) 
    print('\n')
```

```python
max_acc1 = 0
random_s1 = 0

for i in range(200):
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.30, random_state = i)
    model1 = LinearRegression()
    model1.fit(x_train, y_train)
    model1_predict = model1.predict(x_test)
    acc = r2_score(model1_predict, y_test)
    
    if acc > max_acc1:
        max_acc1 = acc
        random_s1 = i
print('Model is giving best accurancy', max_acc1, 'on random state of', random_s1 )

# Model is giving best accurancy 0.9288123541245199 on random state of 115
```

```python
# Model is giving best accuracy at 115 but that is acutally a under fitting model 
# where accuracy of testing > accuracy of training 
# As we can observe in above output 
# at the random state of 38 accuracy of test is almost same as accuracy of train data
```

```python
best_fit_linear = 38  # at this random state accuracy of training and testing data is almost same
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.30, random_state = best_fit_linear)
linear_r = LinearRegression()
linear_r.fit(x_train, y_train)
lr_pred = linear_r.predict(x_test)
lr_pred_train = linear_r.predict(x_train)

Train_accuracy1 = r2_score(lr_pred_train, y_train)
Test_accuracy1 = r2_score(lr_pred, y_test)
print('Training accuracy: ', Train_accuracy1)
print('Testing accuracy: ', Test_accuracy1)
print('______________________________________________')
print('Mean squared error: ', mean_squared_error(lr_pred, y_test) )
print('Mean absolute error: ', mean_absolute_error(lr_pred, y_test) )


# Training accuracy:  0.8856786296702359
# Testing accuracy:  0.8855577065563986
# ______________________________________________
# Mean squared error:  2.63890201982328
# Mean absolute error:  1.226389630358252
```

```python
print('shape of x_test', x_test.shape)
print('shape of x_test', x_train.shape)

print('shape of y_test', y_test.shape)
print('shape of y_train', y_train.shape)

print('shape of x', x.shape)
print('shape of y', y.shape)

# shape of x_test (60, 3)
# shape of x_test (138, 3)
# shape of y_test (60,)
# shape of y_train (138,)
# shape of x (198, 3)
# shape of y (198,)
```

#### Cross Validation for LinearRegression

```python
for value in range (2, 8):
    cv = cross_val_score(linear_r, x, y, cv = value).mean()
    print('cv score is', cv, 'at cross fold value', value, 'with training accurancy of ', Train_accuracy1, 'and testing accuracy of ',Test_accuracy1 )
    
```
